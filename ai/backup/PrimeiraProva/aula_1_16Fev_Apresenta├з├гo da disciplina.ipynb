# Fundamentos e paradigmas da Inteligência Artificial (IA)

Lógica e Raciocínio: A lógica matemática é a base da IA simbólica, que utiliza símbolos e regras para representar o conhecimento e raciocinar sobre ele.

Aprendizado de Máquina (Machine Learning): O aprendizado de máquina é uma abordagem em que os sistemas aprendem a partir de dados, identificando padrões e fazendo previsões sem a necessidade de programação explícita.

Redes Neurais Artificiais: Inspiradas pelo funcionamento do cérebro humano, as redes neurais artificiais são modelos computacionais compostos por neurônios interconectados que aprendem a realizar tarefas através do ajuste dos pesos das conexões.

Raciocínio Probabilístico: Modelos probabilísticos são usados para lidar com a incerteza e a variabilidade nos dados, permitindo que os sistemas façam inferências e tomem decisões baseadas em probabilidades.

Agentes Inteligentes: Os agentes inteligentes são entidades computacionais capazes de perceber o ambiente, tomar decisões e agir de forma autônoma para alcançar seus objetivos.

Processamento de Linguagem Natural (PLN): O PLN envolve o desenvolvimento de sistemas que compreendem e geram linguagem humana, permitindo a interação entre humanos e máquinas através de texto ou voz.

Visão Computacional: A visão computacional capacita os sistemas a interpretar e entender o conteúdo visual de imagens e vídeos, permitindo a detecção de objetos, reconhecimento facial, entre outras aplicações.

Sistemas Especialistas: Os sistemas especialistas são programas de computador que utilizam conhecimento especializado para resolver problemas em um domínio específico, emulando o raciocínio de um especialista humano.

Computação Evolutiva: Inspirada pela teoria da evolução, a computação evolutiva desenvolve algoritmos que buscam soluções ótimas através de processos de seleção natural e reprodução.

Robótica Inteligente: A robótica inteligente combina princípios de IA com sistemas físicos para criar robôs capazes de perceber o ambiente, tomar decisões e interagir com ele de forma autônoma.

# Técnicas de IA aplicadas à solução de problemas

Aprendizado de Máquina<br>
Classificação: Utilizada para determinar a que classe ou categoria os dados pertencem. Exemplo: como detecção de spam em emails, reconhecimento de padrões em imagens médicas para diagnóstico, etc.<br>
Regressão: Aplicada para prever valores numéricos com base em dados históricos, como previsão de preços de imóveis, demanda de produtos, etc.<br>
Agrupamento: Usada para identificar padrões e grupos semelhantes em conjuntos de dados, como segmentação de clientes para campanhas de marketing direcionado.<br>
Aprendizado por reforço: Empregado em sistemas de tomada de decisão sequencial, como jogos de videogame, robótica, controle de processos, etc.

Redes Neurais Artificiais
Redes Neurais Convolucionais: Amplamente utilizadas em visão computacional para tarefas como reconhecimento de objetos, detecção facial, etc.
Redes Neurais Recorrentes: Aplicadas em processamento de linguagem natural (PLN) para tarefas como tradução automática, geração de texto, análise de sentimentos, etc.

Processamento de Linguagem Natural
Análise de Sentimentos: Determinação do sentimento expresso em textos, útil para análise de feedback de clientes, mídias sociais, etc.
Extração de Informações: Identificação e extração de informações relevantes de documentos, como nome de pessoas, locais, datas, etc.
Tradução Automática: Tradução de textos de uma língua para outra, como o Google Translate.
Resposta a Perguntas (Q&A): Resposta automática a perguntas formuladas em linguagem natural, como assistentes virtuais.

Processamento de Imagem e Visão Computacional
Detecção de Objetos: Identificação e localização de objetos em imagens ou vídeos, amplamente utilizado em sistemas de vigilância, veículos autônomos, etc.
Segmentação de Imagens: Divisão de uma imagem em partes ou regiões significativas, como em aplicações médicas para identificar órgãos.
Reconhecimento Facial: Identificação de faces em imagens ou vídeos, aplicado em sistemas de segurança, reconhecimento biométrico, etc.

Algoritmos Evolucionários
Otimização: Utilizados para encontrar soluções ótimas em problemas complexos e multidimensionais, como design de produtos, planejamento logístico, etc.

# Sistemas baseados em conhecimento, planejamento e aprendizagem

1 - Sistemas Baseados em Conhecimento
Os sistemas baseados em conhecimento são projetados para representar e manipular o conhecimento de especialistas humanos em um domínio específico.
Eles usam uma base de conhecimento formalizada, que pode incluir regras, fatos, heurísticas e modelos conceituais.
Esses sistemas são especialmente úteis quando o conhecimento sobre um domínio é bem estruturado e pode ser expresso em termos lógicos ou simbólicos.
Exemplos incluem sistemas especialistas, que são programas de computador que tomam decisões ou fornecem conselhos em um domínio específico, como medicina ou finanças.

2 - Sistemas de Planejamento
Os sistemas de planejamento são projetados para criar sequências de ações para atingir um objetivo específico, levando em consideração restrições e condições do ambiente.
Eles operam em ambientes complexos e dinâmicos, onde múltiplas ações podem ser necessárias para alcançar um objetivo desejado.
Esses sistemas podem usar técnicas como busca heurística, algoritmos de otimização e raciocínio sobre ações possíveis.
Exemplos incluem sistemas de planejamento em robótica, logística, gerenciamento de projetos, entre outros.

3 - Sistemas de Aprendizagem
Os sistemas de aprendizagem são projetados para melhorar seu desempenho com base em experiências anteriores.
Eles podem ser divididos em aprendizado supervisionado, não supervisionado e por reforço, dependendo do tipo de feedback disponível durante o processo de aprendizagem.
Esses sistemas são capazes de extrair padrões e conhecimentos úteis a partir de grandes volumes de dados.
Exemplos incluem sistemas de recomendação, reconhecimento de padrões em imagens, processamento de linguagem natural.

A combinação dessas abordagens pode resultar em sistemas mais robustos e adaptáveis, capazes de lidar com uma variedade de problemas em diferentes domínios. 
Por exemplo, um sistema de diagnóstico médico pode usar um sistema baseado em conhecimento para representar as regras e heurísticas dos médicos especialistas, um sistema de planejamento para determinar a ordem de execução dos testes de diagnóstico e um sistema de aprendizado para melhorar a precisão do diagnóstico com base em dados históricos de pacientes.

Heurística é um termo que se refere a uma técnica ou método utilizado na resolução de problemas que busca uma solução aceitável, mesmo que não seja a mais ótima ou perfeita. Em outras palavras, a heurística é uma abordagem prática e baseada na experiência para encontrar soluções eficientes em situações complexas ou difíceis de resolver de maneira exata.

Características:
Simplicidade: As heurísticas são geralmente fáceis de entender e implementar, sem exigir cálculos ou análises complexas.
Eficiência: Elas são projetadas para encontrar soluções rapidamente, mesmo que não sejam garantidas de serem as melhores possíveis.
Adaptabilidade: As heurísticas podem ser ajustadas ou modificadas conforme necessário para lidar com diferentes contextos ou problemas.
Subjetividade: A aplicação de uma heurística muitas vezes envolve algum grau de julgamento subjetivo por parte do usuário.

Exemplos comuns de heurísticas

Heurística do Vizinho Mais Próximo: Uma abordagem para encontrar soluções aproximadas em problemas de otimização, onde uma decisão é tomada com base na proximidade imediata sem considerar todas as alternativas possíveis.
Heurística do Gradiente Descendente: Utilizada em otimização, esta heurística ajusta, iterativamente, uma solução em direção ao mínimo de uma função de custo, seguindo o gradiente da função.
Heurística de Seleção de Características (Feature Selection): Em problemas de busca, esta heurística envolve priorizar a exploração de características mais promissores com base em alguma métrica mais relevantes ou úteis. (Atributos de predição, preditores ou variáveis independentes ou características ou features; Atributos-alvo ou target )

As heurísticas são amplamente utilizadas em diversos campos, incluindo ciência da computação, inteligência artificial, psicologia, economia e tomada de decisões em geral, como uma maneira de lidar com a complexidade e a incerteza em situações do mundo real.

# Aprendizado de Máquina

### Definido como
>o processo de permitir que um programa aprenda padrões a partir de dados, para que ele possa fazer previsões ou tomar decisões<br>
### Realiza tarefas de
>**Classificação:** o objetivo é categorizar algo em classes, por exemplo, comentários sobre um produto que pode ser “bom”, “ruim” ou “cumpre o que promete”. A saída é categórica ou qualitativa (não numérica).<br>
>**Regressão:** o objetivo é prever um valor numérico, por exemplo, o preço de um imóvel. A saída é um valor contínuo ou quantitativo.<br>
>**Clusterização:** o objetivo é encontrar grupos de dados com características em comum, por exemplo, agrupar clientes por comportamento de compra. Não há as saídas como na regressão e na classificação.<br>

### Fundamentado em dados, algoritmos e modelos
#### Dados em tabelas
>Dados de treinamento incluem rótulos => **Aprendizado supervisionado**<br>
>Dados de treinamento com rótulos ausentes => **Aprendizado não supervisionado**<br>
#### Tarefas de
>Regressão e Classificação => ligadas ao aprendizado supervisionado<br>
>Clusterização => ligada ao aprendizado não supervisionado<br>

### Manipulação dos dados
>pandas

###  Algoritmos
>Sequência de passos ou regras que um computador segue para realizar uma tarefa específica. É uma espécie de receita que orienta o computador sobre como aprender com os dados.<br>
>Algoritmos de aprendizado de máquina podem ser classificados em várias categorias, como: algoritmos de regressão, classificação, agrupamento etc. Cada tipo de algoritmo é projetado para resolver um tipo específico de problema.

### Modelos
>Modelo é uma representação matemática ou computacional de um processo ou sistema que pode aprender a partir de dados.<br>
>O modelo é construído para capturar os padrões, relações e informações presentes nos dados de treinamento, para que possa fazer previsões, tomar decisões ou realizar outras tarefas semelhantes em novos dados.<br>
>Um modelo é o resultado do treinamento (processo de aprendizado) de um algoritmo com dados específicos.<br>

|Algoritmo	|Modelo (algoritmo treinado)|
|------|------|
|Regressão linear: aprendizado supervisionado usado para prever um valor contínuo (numérico) com base em **variáveis independentes**. Por exemplo: prever o preço de uma casa com base em sua área, número de quartos etc. O algoritmo de regressão linear encontra a relação entre essas variáveis e o preço da casa. O preço da casa é a variável dependente, resposta ou target.|Regressão Linear: tenta encontrar a reta que melhor se ajusta aos dados. É o mais simples, o mais **interpretável** e o mais usado. E é o que os alunos menos dão bola. Árvores de decisão, Support Vector Machine (SVM), k-Nearest Neighbors (k-NN)|
|Análise de Componentes Principais (PCA): aprendizado não supervisionado usado para realizar redução de dimensionalidade, processo que transforma um conjunto de variáveis correlacionadas em um conjunto de variáveis não correlacionadas (chamadas componentes principais). Ele é frequentemente usado para visualização de dados e para eliminar redundâncias nos dados.|PCA|
|K-Means: algoritmo de agrupamento que divide um conjunto de dados em grupos (clusters) com base na similaridade.|K-Means: identifica grupos ou clusters nos dados, sem a necessidade de rótulos |

>**Variáveis independentes** são o mesmo que: características, features, atributos de predição e preditores.<br>
>**Modelos interpretáveis:** modelos lineares generalizados - regressão linear e regressão logística, análise discriminante linear, SVMs, árvores de decisão.<br>
>**Menos interpretáveis:** redes neurais, SVMs não lineares, florestas aleatórias.<br>

### Eficácia dos Modelos
>A acurácia da classificação e o score da regressão são avaliados através de métricas, como o **erro médio quadrático** (MSE)<br>
>O erro médio quadrático (MSE) é usado para **ajustar os parâmetros** do modelo para melhorar seu desempenho<br>

#### Diferença de interpretação entre a acurácia da classificação e o score da regressão:
>* Acurácia é a porcentagem de acertos na classificação
>* Score = (variância da média - variância do modelo) / variância da média
>>- Se a variância do modelo for pequena, score é próximo de 1 e o modelo é muito bom
>>- Se a variância do modelo for próxima da variância da média, score é próximo de 0 e o modelo é muito ruim
>>- Se a variância do modelo for maior do que a variância da média, score é negativo e o modelo é um fiasco

### Parâmetros do modelo
> * A Regressão Linear possui 2 parâmetros para ajustar durante o processo de treinamento<br>
> * O Generative Pre-trained Transformer (GPT), desenvolvido pela OpenAI, usado pelo ChatGPT, possui 175 bilhões de parâmetros para ajustar durante o processo de treinamento. O GPT é um exemplo de uma Transformer Neural Network, que é uma arquitetura de Deep Learning projetada para lidar com tarefas de processamento de linguagem natural<br>

# Conceitos de estatística necessários

>Média, Mediana e Moda: Essas são medidas de tendência central que ajudam a entender onde os dados estão centralizados.<br><br>
>Desvio Padrão e Variância: Essas medidas ajudam a entender a dispersão dos dados em torno da média.<br><br>
>Distribuições de Probabilidade: É importante entender diferentes distribuições de probabilidade, como a distribuição normal, e como elas são usadas em modelos de aprendizado de máquina.<br><br>
>Correlação: Ajuda a entender o relacionamento entre diferentes variáveis e pode ser útil para a seleção de features e a compreensão dos dados.<br><br>
>Testes de Hipóteses: Compreender os conceitos básicos de testes de hipóteses pode ser útil para avaliar a significância estatística dos resultados dos modelos de aprendizado de máquina.<br><br>
>Regressão e Análise de Variância (ANOVA): Estes são métodos estatísticos usados para modelar e analisar relacionamentos entre variáveis.<br><br>
>Amostragem e Estimação de Parâmetros: Compreender como amostrar dados e estimar parâmetros populacionais é fundamental para a análise estatística.<br><br>
>Validação de Modelo e Métricas de Avaliação: É importante entender como avaliar e validar modelos de aprendizado de máquina usando diferentes métricas.<br><br>
Pode-se aprender esses conceitos conforme avança no estudo de aprendizado de máquina e não é necessário ser um especialista em estatística desde o início
